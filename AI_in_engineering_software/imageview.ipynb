```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Загрузка изображения
img = cv2.imread('C:/Users/Alex/3D Objects/Total_view_data/CSG_data_PNG/00000001_f.png')

# Преобразование в HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Определение диапазонов цветов
# Более узкие диапазоны для большей чувствительности
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 25])  # Снижение верхней границы для насыщенности

# Маски для каждого цвета
mask_red = cv2.inRange(hsv, lower_red, upper_red)
mask_black = cv2.inRange(hsv, lower_black, upper_black)

# Создание матриц XY
matrix_red = np.zeros((180, 180))
matrix_black = np.zeros((180, 180))

# Заполнение матриц значениями из масок
for i in range(180):
    for j in range(180):
        if mask_red[i, j] == 255:
            matrix_red[i, j] = 1
        if mask_black[i, j] == 255:
            matrix_black[i, j] = 1

# Вывод матриц
plt.subplot(121)
plt.imshow(matrix_red, cmap='gray')
plt.title('Красная матрица')

plt.subplot(122)
plt.imshow(matrix_black, cmap='gray')
plt.title('Черная матрица')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def read_xyz(xyz_file):
    # Read the XYZ file
    vertices = np.loadtxt(xyz_file)
    return vertices

def visualize_xyz(xyz_file):
    # Read vertices from the XYZ file
    vertices = read_xyz(xyz_file)

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Extract coordinates
    x = vertices[:, 0]
    y = vertices[:, 1]
    z = vertices[:, 2]

    # Plot the vertices with smaller size
    ax.scatter(x, y, z, c='r', marker='o', s=1)  # Уменьшить размер точек до 1

    # Set labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Show the plot
    plt.show()

visualize_xyz("output.xyz")

```
