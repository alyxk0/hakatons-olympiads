```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from filterpy.kalman import KalmanFilter
import matplotlib.pyplot as plt

# Read training dataset from csv file
traini_df = pd.read_csv('values.csv', delimiter=';')

X_train = traini_df.iloc[:, :-1]
data_C26 = traini_df.iloc[:, ::8]
data_C15 = traini_df.iloc[:, 1::8]
data_F15 = traini_df.iloc[:, 2::8]
data_F26 = traini_df.iloc[:, 3::8]
data_P15 = traini_df.iloc[:, 4::8]
data_P26 = traini_df.iloc[:, 5::8]
data_O26 = traini_df.iloc[:, 6::8]
data_O15 = traini_df.iloc[:, 7::8]
```

```python
X_train_C26 = data_C26.iloc[:, :-1] 
X_train_C15 = data_C15.iloc[:, :-1] 
X_train_F15 = data_F15.iloc[:, :-1] 
X_train_F26 = data_F26.iloc[:, :-1] 
X_train_P15 = data_P15.iloc[:, :-1] 
X_train_P26 = data_P26.iloc[:, :-1]
X_train_O26 = data_P15.iloc[:, :-1] 
X_train_O15 = data_P26.iloc[:, :-1]

y_train = traini_df.iloc[:, -1] 
```

```python
X_train_C26_mean = np.abs(X_train_C26).mean(axis=1) 
X_train_C15_mean = np.abs(X_train_C15).mean(axis=1)
X_train_F15_mean = np.abs(X_train_F15).mean(axis=1)
X_train_F26_mean = np.abs(X_train_F26).mean(axis=1)
X_train_P15_mean = np.abs(X_train_P15).mean(axis=1)
X_train_P26_mean = np.abs(X_train_P26).mean(axis=1)
X_train_O26_mean = np.abs(X_train_O26).mean(axis=1)
X_train_O15_mean = np.abs(X_train_O15).mean(axis=1)
```

```python
mean_value = np.mean(X_train_C26_mean)
plt.plot(X_train_C26_mean)
plt.title('X_train_C26_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_C15_mean)
plt.plot(X_train_C15_mean)
plt.title('X_train_C15_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_F15_mean)
plt.plot(X_train_F15_mean)
plt.title('X_train_F15_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_F26_mean)
plt.plot(X_train_F26_mean)
plt.title('X_train_F26_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_P15_mean)
plt.plot(X_train_P15_mean)
plt.title('X_train_P15_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_P26_mean)
plt.plot(X_train_P26_mean)
plt.title('X_train_P26_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_O26_mean)
plt.plot(X_train_O26_mean)
plt.title('X_train_O26_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()

mean_value = np.mean(X_train_O15_mean)
plt.plot(X_train_O15_mean)
plt.title('X_train_O15_mean')
plt.xlabel('Sample')
plt.ylabel('Mean amplitude')
plt.axhline(y=mean_value, color='r', linestyle='-')
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier

modelLR = SVC()
modelLD = AdaBoostClassifier()
modelMLP = MLPClassifier()

modelLR.fit(X_train, y_train)
modelLD.fit(X_train, y_train)
modelMLP.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

models = [('SVC', modelLR), ('LD', modelLD), ('MLP', modelMLP)]

for name, model in models:
    scores = cross_val_score(model, X_train, y_train, cv=6)
    print(f'{name}: {scores.mean()}')
```
